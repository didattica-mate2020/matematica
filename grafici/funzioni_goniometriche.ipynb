# Le funzioni goniometriche $\sin$, $\cos$, $\tan$

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
#ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, np.cos(x), label='coseno')  # Plot more data on the axes...
ax.plot(x, np.sin(x), label='seno')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(6, 3), layout='constrained')
#ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, np.tan(x), label='tangente')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
plt.show()
```

## Dipendenza dai parametri

Considerando la funzione 
$$y=A\cdot \sin(\omega x +\phi) +B$$  
analizziamo come varia il grafico al variare dei parametri.

### Pulsazione $\omega$

```python
x = np.linspace(0, 2*np.pi, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
#ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, np.sin(x), label='sin(x)')  # Plot more data on the axes...
ax.plot(x, np.sin(2*x), label='sin(2x)')  # ... and some more.
ax.plot(x, np.sin(3*x), label='sin(3x)')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
plt.show()
```
